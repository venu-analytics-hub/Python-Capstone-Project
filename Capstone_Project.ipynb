```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Create sample data for demonstration purposes (replace with actual if available)
project_data = {
    'ID': ['A001', 'A002', 'A003', 'A004', 'A005'],
    'Project': ['Project1', 'Project2', 'Project3', 'Project4', 'Project5'],
    'Cost': [1000000, np.nan, 850000, np.nan, 680000],
    'Status': ['Finished', 'Ongoing', 'Finished', 'Failed', 'Failed']
}
employee_data = {
    'ID': ['A001', 'A002', 'A003', 'A004', 'A005'],
    'Name': ['John Alter', 'Alice Luxumberg', 'Bob Marley', 'Sophia Lane', 'Amy Johny'],
    'Gender': ['M', 'F', 'M', 'F', 'F'],
    'City': ['London', 'Toronto', 'Berlin', 'Madrid', 'Oslo'],
    'Age': [32, 29, 30, 28, 27]
}
seniority_data = {
    'ID': ['A001', 'A002', 'A003', 'A004', 'A005'],
    'Designation Level': [2, 3, 1, 2, 3]
}

# Convert dictionaries to DataFrames
project_df = pd.DataFrame(project_data)
employee_df = pd.DataFrame(employee_data)
seniority_df = pd.DataFrame(seniority_data)

# Save to CSV files
project_df.to_csv('project.csv', index=False)
employee_df.to_csv('employee.csv', index=False)
seniority_df.to_csv('seniority.csv', index=False)
```

## Task 2: Replace Missing Cost Values with Running Average

```python
# Load CSV
project_df = pd.read_csv('project.csv')

# Replace missing 'Cost' values with running average using a for loop
for i in range(len(project_df)):
    if pd.isna(project_df.loc[i, 'Cost']):
        project_df.loc[i, 'Cost'] = project_df['Cost'][:i].mean()

# View updated DataFrame
project_df
```

## Task 3: Split Name into First and Last Name

```python
# Load employee DataFrame
employee_df = pd.read_csv('employee.csv')

# Split 'Name' column into 'First Name' and 'Last Name'
employee_df[['First Name', 'Last Name']] = employee_df['Name'].str.split(' ', n=1, expand=True)
employee_df.drop('Name', axis=1, inplace=True)
employee_df
```

## Task 4: Join All DataFrames into Final

```python
# Load seniority DataFrame
seniority_df = pd.read_csv('seniority.csv')

# Merge all DataFrames into one
merged_df = pd.merge(project_df, employee_df, on='ID')
Final = pd.merge(merged_df, seniority_df, on='ID')
Final
```

## Task 5: Add Bonus Column (5% of Cost if Project is Finished)

```python
# Add 'Bonus' column
Final['Bonus'] = Final.apply(lambda row: 0.05 * row['Cost'] if row['Status'] == 'Finished' else 0, axis=1)
Final[['ID', 'Cost', 'Status', 'Bonus']]
```

## Task 6: Demote Failed Employees and Remove Level > 4

```python
# Demote designation level by 1 for failed projects
failed_ids = Final[Final['Status'] == 'Failed']['ID'].unique()
Final['Designation Level'] = Final.apply(
    lambda row: row['Designation Level'] + 1 if row['ID'] in failed_ids else row['Designation Level'], axis=1
)

# Remove employees with designation level > 4
Final = Final[Final['Designation Level'] <= 4].reset_index(drop=True)
Final[['ID', 'Status', 'Designation Level']]
```

## Task 7: Prefix Mr./Mrs. to First Name and Drop Gender

```python
# Prefix based on gender
Final['First Name'] = Final.apply(
    lambda row: f"Mr. {row['First Name']}" if row['Gender'] == 'M' else f"Mrs. {row['First Name']}", axis=1
)

# Drop Gender column
Final.drop('Gender', axis=1, inplace=True)
Final[['ID', 'First Name', 'Last Name']]
```

## Task 8: Promote Employees Aged Over 29

```python
# Promote by reducing level by 1 (minimum is 1)
Final['Designation Level'] = Final.apply(
    lambda row: row['Designation Level'] - 1 if row['Age'] > 29 and row['Designation Level'] > 1 else row['Designation Level'],
    axis=1
)
Final[['ID', 'Age', 'Designation Level']]
```

## Task 10: Filter Employees Where City Contains 'o'

```python
# Filter by city name containing 'o'
filtered_employees = Final[Final['City'].str.contains('o', case=False, na=False)]
filtered_employees
```
